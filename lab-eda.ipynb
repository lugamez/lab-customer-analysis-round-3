2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
customer_data[(customer_data['Response'] == 'Yes')].groupby(['Policy Type', 'Gender']).agg({'Total Claim Amount':'mean'}).round(2)
```

Among the people that responded "Yes" to the marketing campaign, we can take the following conclusions:

1. Women with a Corporate Auto or Special Auto policy have a higher claim amount than men with the same one
2. Men with a Personal Auto policy have a higher claim amount than women with the same one
3. There's a 5 point difference in the total claim amount between those women and men that have a Personal Auto policy. For the other policies, the difference increases 5x in comparison with the Personal policy.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# As 631 customers (5% of the dataframe) don't have information about the state, we will fill these rows with the mode (California)

customer_data['State'].mode()
```

```python
# Fill rows from customers with NaN as State

customer_data['State'].fillna('California', inplace=True)
```

```python
# Total number of customer who have policies in each state

customer_per_state = customer_data.groupby(['State']).agg({'Customer':'count'})
display(customer_per_state)
```

```python
# Results filtered to only include states where there are more than 500 customers

customer_per_state[(customer_per_state['Customer'] > 500)]
```

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
# Make a copy of the dataframe

customer_data_copy = customer_data.copy()
```

```python
# Extract the month from the column "Efective To Date"

customer_data_copy[['Policy - Month Eff', 'Policy - Day Eff', 'Policy - Year Eff']] = customer_data_copy['Effective To Date'].str.split('/', expand=True)
display(customer_data_copy)
```

```python
# Create pivot

policies_state_month = customer_data_copy.pivot_table(index='State', columns='Policy - Month Eff', values='Number of Policies', aggfunc='count')
display(policies_state_month)
```

6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
# Sort values in descending order

top_states = policies_state_month.sort_values(by=['1','2'], ascending=False)
display(top_states)
```

```python
# Show 3 states with the highest number of policies sold

top_states.iloc[:3]
```
